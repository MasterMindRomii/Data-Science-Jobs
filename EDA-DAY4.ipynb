# Data Visualization

```python
df.columns
```

```python
import matplotlib.pyplot as plt
```

```python
# Create a histogram of the column
plt.hist(df['work_year'])
plt.xlabel=('experience_level')
plt.ylabel=('job_title')
plt.title('Distribution of job vs Exp')
plt.show()
```

```python
plt.hist(df['salary_in_usd'])
plt.xlable=('Job_title')
plt.ylabel=('salary')
plt.title('Job_Title Vs Salary')
plt.show()
```

```python
# Create a scatter plot of the columns
plt.scatter(df['salary'], df['salary_in_usd'])
plt.xlabel=('salary')
plt.ylabel=('salary_in_usd')
plt.title('Scatter Plot of Salary Vs. Salary in USD')
plt.show()
```

```python
plt.scatter(df['company_location'],df['company_size'])
plt.xlabel=('company_location')
plt.ylabel=('company_size')
plt.title('Scatter Plot of Company_location Vs. Company_size')
plt.show()
```
